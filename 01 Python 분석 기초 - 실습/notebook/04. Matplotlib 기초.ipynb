```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## 1. Bar chart:  유형별 통계치 요약

- 가로축에는 명목형변수(a반,b반,c반에는 순서가 없음)
- 세로축에는 수치형변수 <br><br>

- 명목형변수가 1개면 빈도(도수분포표)그린후  bar차트 
- 명목형변수가 2개면 분할표(테이블)를 만듬
- 남녀(변수이름이 아니라 값의 이름)별 ab반에 따른 명수 <br><br>

- 막대그래프에는 적어도 명목형 변수가 1개는 있어야함
- 수치형변수는 없어도 됨 <br><br>

- 막대로 가장 효과적으로 나타내는건 도수분포표와 분할표임 <br><br>

- bar의 높이를 카운트가 아닌 비율로 나타내면 파이차트 <br><br>


```python
x = np.array(['Q1', 'Q2', 'Q3','Q4'])
y = np.array([ 234.0, 254.7, 144.6, 317.6])
plt.bar(x,y,color = 'purple')
plt.title('BAR CHART')
plt.show()
```

## 2. Histogram: 치우친 정도, 외상치관리, 다른 데이터 개입여부 알기 위함

수치형변수이고 연속형변수 1개 가지고 그림

수치와 명목을 조합해서 히스토그램 그리려면?<br>

1. 세로방향에 남자유형,여자유형해서 가로축에 구간맞춰서 비교<br><br>

2. 똑같은 프레임안에 색깔을 구분하고, 투명도를 구별하는거<br>
투명도=알파값(겹친부분은 겹쳤다고,안겹친부분은 안겹쳤다는걸 알아보기 위해서)

```python
x = np.random.randn(10000)
plt.hist(x,bins=20,color='green',density=True)
plt.title('HISTOGRAM')
plt.show()
```

```python
x = np.random.randn(10000)
plt.hist(x,bins=100,color='blue',density=False,alpha=0.3)
plt.title('HISTOGRAM')
plt.show()
```

## 3. Box plot: 전체분포와 아웃라이어(결측치) 확인

- 수치형변수1개면 박스플롯1개(통계와 아웃라이어의 목적)
- 수치형과 명목형이라면 박스플롯 여러개(유형별 통계적 분포, 아웃라이어, 비교의 목적)<br><br>

- 박스가 의미하는 것?
- 분위수보다 확률값이 크거나 작을 [위치]
- 가설검정 신뢰구간에서 분위수가 매우 중요
- 25,50,75분위수(구간이 4개라 4분위수)<br><br>

- 분위수가지고 할 수 있는것
- 등급을 나누면 개별적으로 관리할 필요없음
- 등급별로 관리하면 됨<br><br>

- 등급나누는것=군집화
- 군집화중 가장 단순한게 분위수로 나누는것<br><br>


```python
x = np.random.randn(10000)*10+3
y = np.random.randn(10000)*10+5
z = np.random.randn(10000)*10+1
plt.boxplot([x,y,z],0)
plt.title('BOX PLOTS')
plt.show()
```

```python
plt.boxplot([x,y,z],0,vert=False)
plt.title('BOX PLOTS')
plt.show()
```

## 4. Line plot: 추세를 보여주기 위함

- 가로축, 세로축 모두 수치형 변수 <br><br>
- 시계열 시각화에 사용

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='orange',linewidth=2)               # linestyle = '-'
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='blue',linestyle=':',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='red',linestyle='-.',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='green',linestyle='--',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='purple',linestyle='steps',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='purple',linestyle='-',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.xlim([0,5])                                     # 가로 방향 범위.
plt.ylim([-2,+2])                                   # 세로 방향 범위.
plt.show()
```

## 5.1 Scatter plot 1: 가로축과 세로축 변수사이의 관계가 중요, 상관계수

- 상관계수(x와 y사이에 선형관계가 있느냐)
- 상관계수는 2개의 변수가 있어야 함<br><br>

- 본인자신에 대한 상관성은 의미없음. 항상 1임
- -1~+1값을 취함<br><br>

- 상관성이 강하다는 것은 상관계수가 절대값 1에 가깝다
- -1이면 음의 상관성이 강하고 +1이면 양의 상관성이 강하다
- 0에 가까우면 상관성이 약하다

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='red',marker='o',linestyle='none',markersize=5, alpha=0.5)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='#FF8C00',marker='o',linestyle='none',markersize=5, alpha=0.5)     # RGB color.
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,marker='o',linestyle='none',markersize=15,markerfacecolor='yellow',markeredgecolor='purple',markeredgewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```
