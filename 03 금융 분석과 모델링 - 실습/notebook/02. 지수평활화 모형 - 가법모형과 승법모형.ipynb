## 지수평활화 모형

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import warnings
warnings.filterwarnings("ignore")  # 성가신 warning을 꺼준다.
%matplotlib inline
```

```python
os.chdir(r"C:\Users\Gram\Desktop\myPyCode\03 금융 분석과 모델링 - 실습\data")
```

```python
df = pd.read_csv('data_time_series.csv', header='infer',encoding = 'latin1')
```

```python
df.plot()
plt.show()
```

### 1) 가법모형

```python
data = df['Value']  #지수평활화 (expo~)                #가법
model = ExponentialSmoothing(data.iloc[0:35], seasonal='add',seasonal_periods=6).fit()   
pred = model.predict(start= 36, end=47)  
plt.plot(data,color='blue')
plt.plot(pred,color='red')
plt.show()
```

### 2) 승법모형

```python
data = df['Value']                                     #승법
model = ExponentialSmoothing(data.iloc[0:35], seasonal='mul',seasonal_periods=6).fit()   
pred = model.predict(start= 36, end=47)
plt.plot(data,color='blue')
plt.plot(pred,color='red')
plt.show()
```
