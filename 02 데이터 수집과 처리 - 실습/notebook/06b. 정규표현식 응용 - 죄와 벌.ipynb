## 정규표현식 응용

```python
import re
import os
```

```python
os.chdir(r"C:\Users\Gram\Desktop\myPyCode\02  데이터 수집과 처리 - 실습\data")
f = open("./book_crime_and_punishment.txt",'r',encoding='UTF8')
book = f.readlines()
f.close()
```

```python
def grep(reg_exp,my_text):    #grep은 정규표현식을 넣어주고 리스트를 넣어주면 리스트에서 정규표현식이 발견되면 가져와주는거
    result = []
    reg_exp_compiled = re.compile(reg_exp)  #정규표현식 컴파일
    for i in range(len(my_text)):           #패턴이 발견된 문자들을 모아두었다가 출력시킴
        if reg_exp_compiled.search(my_text[i]):
            result.append(my_text[i])
    return result
```

### 1. 간단한 테스트

```python
grep('hate', ['I love you', 'I hate you', 'I like you'])   
```

```python
grep("^[0-9]", ["2018 is this", "you are number 212", "my age is 21"])   
```

```python
grep("r{3}.", ["Dog barks", "Cat purrrs", "Brroong"])   # r이 3번반복되는 패턴
```

### 2. "죄와 벌"에서 발췌

```python
grep("^This",book)            # "This"로 시작하는 문장 가져오기
```

```python
grep("end[.]$",book)          # [.]나  \.하면 메타문자.이 아닌 마침표를 의미 =>  end.으로 끝나는 문장들 가져오기
```

```python
grep("^[Ii]",book)            # 대문자 I 또는 소문자 i로 시작하는 문장들
```

```python
grep("^[0-9]",book)           # 숫자로 시작.    
```

```python
grep("b.d",book)              # 아무런 문자 1회.
```

```python
grep("b...d",book)            # 아무런 문자 3회.
```

```python
grep("a*d",book)              # 'a' 0회 이상 반복.
```

```python
grep("a+d",book)              # 'a' 1회 이상 반복.
```

```python
grep("r{2}.d",book)           # 'r' 2회 반복.
```

```python
grep("r{1,2}.d",book)         # 'r' 1~2회 반복.
```

```python
grep("r?ed",book)             # 'r' 0~1회.
```

```python
grep("love|hate",book)        # "love" or "hate".
```

```python
len(grep("love",book))        # book안에 love라는 문장이 몇번 나오는지
```

```python
len(grep("love",book))/len(book)  # 전체 book에서 love가 나오는 비중
```
