## 실제 사이트 크롤링 - 네이버 뉴스

```python
import bs4
import re
import requests as rq
```

### 1. 사이트에 접속하고 탐색해 본다:

1. 먼저 다음 사이트에 들어가서 살펴본다:  [네이버 뉴스](https://news.naver.com)
2. 크롬에서 3점 메뉴 버튼을 누르고 **도구 더보기** ==> **개발자 도구**를 선택해서 개발자 도구를 연다.
3. 개발자 도구 왼쪽 상단의 화살표 버튼을 눌러서 사이트 일부를 서택하고 해당 HTML코드를 볼 수 있다. "[CTRL] + [SHIFT] + C" 조합으로 대체할 수 있다.

```python
# 웹사이트의 URL에 접속.
res = rq.get("https://news.naver.com")     #res=rq.get("주소"): 데이터가져올 사이트 주소
```

```python
# status_code가 200이면 OK.
# status_code가 4xx이면 접속 오류.
res.status_code
```

```python
print(res.text)     # 데이터 전체를 출력(내용을 텍스트로 보여줌)-아직은 파싱이 필요한 상태
                    # print(res.content)와 비교.
```

```python
print(res.content)     # 데이터 전체를 출력(내용을 텍스트로 보여줌)-아직은 파싱이 필요한 상태
```

```python
my_html = res.text      #res.text를 변수 my_html에 입력
soup = bs4.BeautifulSoup(my_html, 'html.parser')        #soup = bs4.BeautifulSoup(res.text,'html.parser') 
                                                        #html을 해석해주는 parser3가지 중 우리는 'html.parser'사용
print(soup.prettify()) #BeautifulSoup 객체 반환. 
```

```python
# 모든 div 태그의 attr을 출력해 본다. 
x=soup.find_all('div')    #soup.find_all('div'): 모든 div태그 출력
n = len(x)      #길이(원소개수)
for i in range(n):
    print(x[i].get('class'))
```

### 2. 헤드라인 가져오기:

#### 1) 먼저 ul 태그로 정리된 헤드라인 리스트 클래스를 가져와 본다.

```python
res = soup.find_all('ul',class_='hdline_article_list')    #ul이면서 헤드라인 article_list인거
for x in res: 
    print(x.contents)
```

#### 2) 위에서 보았을 때 div 태그안에 있는 헤드라인 기사 타이틀을 가져와 본다.

```python
res = soup.find_all('div',class_='hdline_article_tit')   #div면서 ~
for x in res:
    print(x.text.strip())
```

#### 3) 링크 타이틀 중에서 볼드 처리 뉴스 타이틀과 하이퍼링크 가져오기

```python
res_a = soup.find_all('a', class_=re.compile('clicks'))      # "clicks" 패턴 있는 클래스.  #정규표현식
n_count = 0
news = []  #뉴스라는 리스트만듬
for x in res_a:                          
    res = x.find('strong')                                   # 다시 검색해 볼드처리된거만 뽑는다
    if res:                                                  # None이 아닌 경우만.
        an_article = res.text
        if (len(an_article) > 10):                           # 길이가 10 이상인 경우만.
            n_count += 1
            print(n_count, end = " : ")
            print(an_article, end = " => ")
            print(x.get('href'))
            news += [(an_article, x.get("href"))]   #리스트안에 튜플 , 하이퍼링크
```

### 3. 하이퍼링크 따라서 뉴스 기사 가져오기:

```python
for news_title, hlink in news:
        res = rq.get(hlink)   #하이퍼링크따라서 get(가준다)
        my_html = res.text
        soup = bs4.BeautifulSoup(my_html, 'html.parser')  
        body = soup.find(id='articleBodyContents')            #하이퍼링크따라가보니 article~이런게 있음
        if (body):                                            #None이 아닌 경우만 출력.
            print(news_title, end=" : ")
            print(body.text.strip()[63:])                     #서두의 불필요 부분 제거.      
            print("-"*100+"\n")                               #모든 헤드라인 뉴스기사를 다 정리해둔 결과가 나옴
```
