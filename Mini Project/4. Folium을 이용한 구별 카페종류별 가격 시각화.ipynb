```python
#데이터파일 path 설정,
os.chdir(r"C:\Users\Gram\Desktop\구별 프랜차이즈, 개인 카페 가격 비교\1.3 구별 카페 크롤링 결과(결측치 처리 완료)")
file_list = os.listdir()
my_excel_files =[]
for x in file_list:
    if x[-4:] == 'xlsx':
        my_excel_files += [x]
file_list
```

```python
# 읽어 들여서 하나의 데이터 프레임으로 통합한다.
n_excels = len(my_excel_files)
for i in range(n_excels):  
    a_df = pd.read_excel(my_excel_files[i], na_values = [' ', '-'])  #header=2는 첫번째두번째행은 필요없는 정보이기 때문이다.
    if i == 0:
        df = a_df
    else:
        df = pd.concat([df,a_df], axis=0, ignore_index=True)
cafe_seoul=df
header = df.columns
cafe_seoul=df.loc[:,(header=='지역') | (header=='구분') | (header=='가격')]
```

```python
#구 목록
tot=['강남구','강동구','강서구','강북구','관악구',
     '광진구','구로구','금천구','노원구','동대문구',
     '도봉구','동작구','마포구','서대문구','성동구',
     '성북구','서초구','송파구','영등포구','용산구',
     '양천구','은평구','종로구','중구','중랑구']
```

```python
#값을 받을 변수
price=[]

for gu in tot:
    test1_f = cafe_seoul.가격[(cafe_seoul['지역']==gu) & (cafe_seoul['구분']=='F')].mean()
    test1_p = cafe_seoul.가격[(cafe_seoul['지역']==gu) & (cafe_seoul['구분']=='P')].mean()
    test1_r = test1_f - test1_p
    price.append(test1_r)   # 가격 입력
```

```python
#dataframe선언   
test = {"지역":tot, "값": price}
df_test=pd.DataFrame(test,columns=['지역','값'])   
```

```python
#json 데이터 파일에 맞춰주기 위해 index 설정
df_test = df_test.set_index('지역')
```

```python
#df_test 값 column에 따른 값 선언 
map = folium.Map(location=[37.5502, 126.982], zoom_start=11,
                 tiles='cartodbpositron')
map.choropleth(geo_data = geo_str,
               data = df_test['값'],
               columns = [df_test.index, df_test['값']],
               fill_color = 'YlOrRd', #puRd, YlGnBu
               key_on = 'feature.id')
map
```
