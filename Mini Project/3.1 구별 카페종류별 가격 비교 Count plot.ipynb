```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
%matplotlib inline
warnings.filterwarnings('ignore') 
%matplotlib inline
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
# 엑셀 파일 이름을 모아둔다.
os.chdir(r'C:\Users\Gram\Desktop\myPyCode\크롤링을 활용한 구별 카페가격 비교\구별 엑셀파일(결측치 처리후)')
file_list = os.listdir()
my_excel_files =[]
for x in file_list:
    if x[-4:] == 'xlsx':
        my_excel_files += [x]
```

```python
# 읽어 들여서 하나의 데이터 프레임으로 통합한다.
n_excels = len(my_excel_files)
for i in range(n_excels):  
    a_df = pd.read_excel(my_excel_files[i], na_values = [' ', '-'])  #header=2는 첫번째두번째행은 필요없는 정보이기 때문이다.
    if i == 0:
        df = a_df
    else:
        df = pd.concat([df,a_df], axis=0, ignore_index=True)
```

```python
fig = plt.subplots(figsize=(70,30))
sns.set(rc={"font.size":20})
sns.countplot(x='지역', data=df, hue='구분')
sns.set(font_scale=4)
#plt.savefig('fig1.png', dpi=300)  #저장
plt.show()
```
